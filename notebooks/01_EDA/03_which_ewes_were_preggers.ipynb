```python
import pandas as pd
from definitions import RAW_DATA_DIR

filtered_path = RAW_DATA_DIR / "filtered" / "disco_baa_01_filtered_masterfile.csv"

df = pd.read_csv(filtered_path)
```

```python
# find reproductive-related columns
keywords = [
    "preg",
    "scan",
    "lamb",
    "birth",
    "wet",
    "dry",
    "join",
    "mark",
    "twin",
    "trip",
    "fert",
    "repro",
]

repro_cols = [c for c in df.columns if any(k in c.lower() for k in keywords)]
non_null_counts = df[repro_cols].notna().sum().sort_values(ascending=False)

non_null_counts
```

```python
# print the 2024 reproductive columns only
cols_2024 = [c for c in repro_cols if "2024" in c]
_ = [print(c) for c in cols_2024]
```
