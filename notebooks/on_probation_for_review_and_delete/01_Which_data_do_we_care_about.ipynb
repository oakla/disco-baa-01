```python
import os
import pandas as pd
import numpy as np
from pathlib import Path

from definitions import RAW_DATA_DIR
```

```python
df = pd.read_csv(RAW_DATA_DIR / "Heat Stress Masterfile May 2024 - RF Ewe.ram data.csv")
df.head()
```

### Rows
What we know we can get rid of:
- Rows where "sex" is 'M'

```python
df = df[~((df['sex'] == 'M'))]
```

### Columns
Columns we know we need:
- "EID"
- 'EID VID'
- "VID"
- "Preg scan 2024"

Let's find out which other columns might be interesting:
- '2024' columns
- reproductive related columns

```python
# find reproductive-related columns
keywords = [
    "preg",
    "scan",
    "lamb",
    "birth",
    "wet",
    "dry",
    "join",
    "mark",
    "twin",
    "trip",
    "fert",
    "repro",
]
repro_cols = [c for c in df.columns if any(k in c.lower() for k in keywords)]
non_null_counts = df[repro_cols].notna().sum().sort_values(ascending=False)
non_null_counts
```

```python
# find '2024' columns
cols_2024 = [c for c in df.columns if '2024' in c]
non_null_counts_2024 = df[cols_2024].notna().sum().sort_values(ascending=False)
non_null_counts_2024
```
